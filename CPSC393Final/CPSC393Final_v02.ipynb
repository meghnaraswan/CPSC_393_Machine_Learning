```python
years_2010_to_2019_df = pd.read_csv("/Users/iman/CPSC_Courses/CPSC393/CPSC393Final/Data/2010_to_2019.csv")
years_2010_to_2019_df.head()
```

```python
years_2010_to_2019_df.shape
```

```python
# creating vocabulary from training data
years_2010_to_2019_df['lyrics'] = years_2010_to_2019_df['lyrics'].str.split()
vocabulary = []
for lyrics in years_2010_to_2019_df['lyrics']:
   for word in lyrics:
      vocabulary.append(word)
vocabulary = list(set(vocabulary)) # only count the number of unique words
print(len(vocabulary))
```

```python
# each word in each sms_message will be an x-value
word_counts_per_lyrics = {unique_word: [0] * len(years_2010_to_2019_df['lyrics']) for unique_word in vocabulary}

for index, lyrics in enumerate(years_2010_to_2019_df['lyrics']):
   for word in lyrics:
      word_counts_per_lyrics[word][index] += 0
word_counts = pd.DataFrame(word_counts_per_lyrics)
word_counts
```

```python
df_new = pd.concat([years_2010_to_2019_df, word_counts], axis=1) #combine data frames
df_new.head()
```
