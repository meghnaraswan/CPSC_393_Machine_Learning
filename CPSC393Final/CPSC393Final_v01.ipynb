```python
# creating vocabulary from training data
tracks_slang_df['lyrics'] = tracks_slang_df['lyrics'].str.split()
vocabulary = []
for lyrics in tracks_slang_df['lyrics']:
   for word in lyrics:
      vocabulary.append(word)
vocabulary = list(set(vocabulary)) # only count the number of unique words
print(len(vocabulary))
```

```python
# each word in each sms_message will be an x-value
word_counts_per_lyrics = {unique_word: [0] * len(tracks_slang_df['lyrics']) for unique_word in vocabulary}

for index, lyrics in enumerate(tracks_slang_df['lyrics']):
   for word in lyrics:
      word_counts_per_lyrics[word][index] += 0
word_counts = pd.DataFrame(word_counts_per_lyrics)
word_counts
```
